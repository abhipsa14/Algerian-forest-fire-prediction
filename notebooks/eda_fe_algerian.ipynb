Algerian forest fires dataset
The dataset includes 244 instances that regroup a data of two region of algeria, namely the bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

Data cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Region']]=df[['Region']].astype('int')
```

```python
df.head()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.shape
```

```python
## removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
df.shape
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
## dropping indexed 122
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

changes the required columns as integer data type

```python
df.columns
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype('int')
```

```python
df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI', 'Classes']]=df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI', 'Classes']].astype('object')
```

```python
df.info()
```

```python
df.head()
```

Changing other columns to floats datatype

```python
objects=[features for features in df.columns if df[features].dtype=='object']
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype('float')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

Exploratory data analysis

```python
## drop day, month, year
df_copy = df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
## categories in classes
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes'].unique()
```

```python
## encoding of the categories in classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
## plot density for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
## percentage of pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
classlabels = ['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie chart of classes')
plt.show()
```

Correlation

```python
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr(),annot=True)
```

```python
# box plots
sns.boxplot(df['FWI'],color='green')
```

```python
df.head()
```

```python
## monthly fire analysis
dftemp = df.loc[df['Region'] == 1]
fig, ax = plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=dftemp, ax=ax)
ax.set_ylabel('Number of Fires', weight='bold')
ax.set_xlabel('Months', weight='bold')
ax.set_title("Fire Analysis of Sidi- Bel Regions", weight='bold')
plt.show()
```

```python
## Monthly Fire Analysis
dftemp=df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Brjaia Regions",weight='bold')
```
